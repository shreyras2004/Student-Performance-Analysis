# Student Performance Analysis

# Importing Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Loading Dataset

```python
df = pd.read_csv("student-mat.csv", header=None)

df = df[0].str.split(";", expand=True)

df.columns = df.iloc[0]

df = df[1:]

df = df.reset_index(drop=True)

df = df.replace('"', '', regex=True)

df["G3"] = pd.to_numeric(df["G3"])

df["studytime"] = pd.to_numeric(df["studytime"])

df.head()
```

# Data Exploration

```python
print("Dataset Shape:")
print(df.shape)

print("\nMissing Values:")
print(df.isnull().sum())
```

# Average Final Grade Analysis

```python
average_grade = df["G3"].mean()

print("Average Final Grade (G3):")
print(average_grade)
```

# Students Scoring Above 15

```python
high_scorers = df[df["G3"] > 15]

print("Students scoring above 15:")
print(len(high_scorers))
```

# Correlation Between Study Time and Grades

```python
correlation = df["studytime"].corr(df["G3"])

print("Correlation between study time and grades:")
print(correlation)
```

# Gender-wise Performance Analysis

```python
gender_avg = df.groupby("sex")["G3"].mean()

print("Average Score by Gender:")
print(gender_avg)
```

# Histogram of Final Grades

```python
plt.figure(figsize=(8,5))

plt.hist(df["G3"], bins=10, edgecolor='black')

plt.title("Distribution of Final Grades")

plt.xlabel("Final Grade")

plt.ylabel("Number of Students")

plt.savefig("histogram.png")

plt.show()
```

# Scatter Plot: Study Time vs Grades

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x=df["studytime"], y=df["G3"])

plt.title("Study Time vs Final Grade")

plt.xlabel("Study Time")

plt.ylabel("Final Grade")

plt.savefig("scatterplot.png")

plt.show()
```

# Bar Chart: Gender-wise Average Score

```python
plt.figure(figsize=(6,5))

sns.barplot(x=gender_avg.index, y=gender_avg.values)

plt.title("Average Score by Gender")

plt.xlabel("Gender")

plt.ylabel("Average G3 Score")

plt.savefig("barchart.png")

plt.show()
```

# Conclusion

The dataset was successfully cleaned, analyzed, and visualized using Python libraries such as Pandas, Matplotlib, and Seaborn.

The analysis showed that the average student grade was around 10.41. A weak positive correlation was found between study time and grades. Male students performed slightly better on average in this dataset.
